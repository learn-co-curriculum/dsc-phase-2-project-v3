```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from scipy import stats
```

```python
df3 = pd.read_csv("df3.csv")
```

Looking at directors and return percentages

```python
df3['return_perc_v2'].describe()
```

```python
df4 = df3.groupby('primary_name').agg({'count': 'sum', 'return_perc_v2': 'median'}).sort_values('return_perc_v2', ascending = False)
```

```python
df4 = df4.reset_index()
```

```python
df4.to_csv('df4.csv')
```

```python
df4 = df4[ (df4['count'] >= 7) & (df4['return_perc_v2'] > 0) ] # filtering out directors who made 7 or more movies 
```
